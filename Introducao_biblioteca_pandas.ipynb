# **Python para An√°lise de dados(Pandas)**



```python
# importando a biblioteca pandas
import pandas as pd
```

```python
# ler o dataset
df = pd.read_csv('Gapminder.csv', sep=';')
```

```python
# visualizanddo as 5 primeiras linhas
df.head()
```

```python
# Renomear as colunas
df = df.rename(columns={'country':'Pais', 'continent':'Continente', 'year':'Ano',
          'lifeExp':'Expectativa de vida', 'pop':'Pop Total', 'gdpPercap':'PIB'})
df.head()
```

```python
# Quantas linhas e colunas tem no nosso dataframe
df.shape
```

```python
# Verificar as colunas do nosso dataframe
df.columns
```

```python
# Como saber o tipo de cada coluna 
df.dtypes
```

```python
# retornar as ultimas linhas do dataframe
df.tail(3) # o tres √© o numero de linhas a mostrar
```

```python
# Descrever o nossa dataframe
df.describe()
```

```python
# Valore unicos da coluna continente no dataframe
df['Continente'].unique()
```

```python
# para filtrar somente o continente oceania, usando o (loc)
Oceania = df.loc[df['Continente'] == 'Oceania']
Oceania.head()
```

```python
# Vai agrupar e informa quanto pais tem em cada continente
df.groupby('Continente')['Pais'].nunique()
```

```python
# A espequitativa de vida saber a media para cada ano
df.groupby('Ano')['Expectativa de vida'].mean()
```

```python
# Media do PIB
df['PIB'].mean()
```

```python
# Soma do PIB
df['PIB'].sum()
```
